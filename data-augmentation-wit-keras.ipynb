```python
# Linear Algebra
import numpy as np 
# Data processing
import pandas as pd
# Data Visualization
import matplotlib.pyplot as plt
# Stats Visualization
import seaborn as sns
```

### Import Keras Libraries

```python
# Image Generator
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

### Data Generator

```python
# Let's storage it
datagen = ImageDataGenerator(rotation_range=40,
                             # Position change
                             width_shift_range=0.2,
                             # Height change
                             height_shift_range=0.2,
                             # zoom it
                             zoom_range=0.2,
                             # Allowing Mirror type effect
                             horizontal_flip=True,
                             # Filling Image
                             fill_mode='nearest',
                             # Changing Brightness
                             brightness_range=[0.4,1.5])
```

### Array/Image Based Generator

```python
# Storing & loading image
img = load_img('/kaggle/input/joaquin/DSC3359_2_25.jpg')
# Let's pass it into array with Keras
x = img_to_array(img)
# Shape information
print(x.shape)
# Reshaping
x = x.reshape((1,)+ x.shape)
print(x.shape)
```

### Data Generator Based on Directory

```python
# Let's iterate on the Image Generator
i = 0 
for batch in datagen.flow(x, batch_size=1):
    plt.figure(i)
    imgplot = plt.imshow(array_to_img(batch[0]))
    i += 1
    if i % 10 == 0 :
        break
plt.show()    
```

```python
# Let's store a CNN with a specific directory
train_generator = datagen.flow_from_directory('/kaggle/input/cnn-data-sources/cats_and_dogs/train',
                                              # Image due
                                              target_size=(150,150),
                                              # Group set-up
                                              batch_size=32,
                                              # Cats & Dogs model
                                              class_mode='binary')
```
