# Kernels Application Management

### Import Libraries

```python
# Linear algebra
import numpy as np 
# Data processing,
import pandas as pd
# Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
# Scipy
import scipy.ndimage as nd
#
from skimage import io, color
```

### Loading Images

```python
# Lets load an image from the dataset
img = io.imread('/kaggle/input/joaquin/DSC3359_2_25.jpg')
# If possible & there is no conflict, let's switch into gray-scale
img_gray = color.rgb2gray(img)
```

* Gray scale will required just one channel to perform, while color images will required 3 channels. it's for this reason that i decided to switch the photo into gray

```python
img_gray.shape
```

Height and width are set on (428, 437). Since it did not shown any additional argument information. It is assumed that it has a single channel.

```python
# Let's visualize this image
plt.imshow(img_gray)
plt.show()
```

### Kernel Set-Up applying filters

```python
kernel = np.array([[-1,0,1],
                   [-1,0,1],
                   [-1,0,1]])
```

```python
kernel
```

### Convolution Application

```python
# Let's now apply Scipy-ndimage & bring the Kernel previously set-up
img_new = nd.convolve(img_gray, kernel)
```

```python
fig, axes = plt.subplots(1,2, figsize = (15,10))
axes[0].imshow(img_gray, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(img_new, cmap=plt.cm.gray)
axes[1].set_title('Convolution')
axes[1].axis('off')
```
